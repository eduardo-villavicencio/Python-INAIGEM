```python
#importar librerias
from datetime import datetime,timedelta
from dateutil.relativedelta import relativedelta
#Datos de entrada
Ano1=2016
mes1=1
Ano2=2019
mes2=12
#acomodo de datos
fecha1 = datetime(Ano1,mes1,1)
meses1=(abs(Ano2-Ano1)*12)+(abs(mes2-mes1)+1)
meses=range(meses1)
Tiempo1=[]

for dato in meses:
    month = relativedelta(months=dato)
    hoy= fecha1 + month
    Tiempo1.append(str(hoy))
```

```python
#Extrae la informacion de las fechas generadas para el periodo de estudio
H1=[]
for anos in Tiempo1:
    L=anos[:4]+anos[5:7]+anos[8:10]
    H1.append(L)  
#print (L[4:6])
```
